```python
import pandas as pd
import math
from bokeh.palettes import Dark2_5 as palette
import itertools
from bokeh.plotting import figure, output_file, show
```

```python
from bokeh.io import output_file, show
from bokeh.plotting import figure

output_file("bars.html")

# x-coordinates to be plotted
programs = list(dat1['Program'])
   
# x-coordinates of the top edges
top = list(dat1.sum(axis = 1, skipna = True)/(len(dat1.columns)-1))

graph = figure(x_range=programs, height=500, title="Bar Chart",
           toolbar_location=None, tools="")

graph.vbar(x=programs, top=top, width=0.9)

graph.xgrid.grid_line_color = None
graph.y_range.start = 0

show(graph)
```

```python
# file to save the model 
output_file("pie_chart.html") 
           
# instantiating the figure object 
graph = figure(title = "Bokeh Pie Chart") 
  
# name of the sectors
programs = list(dat1['Program'])
  
dat1['Tot \'20'] = dat1['Spring \'20'] + dat1['Fall \'20']
percentages = list((dat1['Tot \'20'] / dat1['Tot \'20'].sum()) * 100)
  
# converting into radians
radians = [math.radians((percent / 100) * 360) for percent in percentages]
  
# starting angle values
start_angle = [math.radians(0)]
prev = start_angle[0]
for i in radians[:-1]:
    start_angle.append(i + prev)
    prev = i + prev
  #ending angle values
end_angle = start_angle[1:] + [math.radians(0)]
  
# center of the pie chart
x = 0
y = 0
  
# radius of the glyphs
radius = 1
  
# create a color iterator
colors = itertools.cycle(palette)    

for i,color in zip(range(len(programs)), colors):
    graph.wedge(x, y, radius,
                start_angle = start_angle[i],
                end_angle = end_angle[i],
                color = color,
                legend = programs[i])
  # displaying the graph
show(graph)
```
