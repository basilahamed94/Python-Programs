```python
import pandas as pd
def main():
    data = pd.read_csv(r'C:\Users\User 8\Downloads\sampledata.csv')
    a = 'y'
    while(a.upper()=='Y'):
        print('\n1. Create')
        print('2. Read')
        print('3. Update')
        print('4. Delete')
        print('5. Exit')
        c = int(input('\nEnter your choice: '))
        if c==1: data = create(data)
        elif c==2: data = read(data)
        elif c==3: data = update(data)
        elif c==4: data = delete(data)
        elif c==5: 
            print('Exiting...')
            break
        else: 
            print('Wrong Choice')
            a=='y'
            continue
            
def create(data):
    print('Enter the values')
    task_id = input('Task ID :')
    priority= input('Priority :')
    category= input('Category :')
    details= input('Details:')
    doe = input('Date of Entry:')
    deadline = input('Deadline:')
    complete = input('Complete:')
    data.loc[len(data.index)] = [task_id,priority,category,details,doe,deadline,complete]
    print('\n',data.iloc[len(data.index)-1])
    print('Row added\n')
    return data

def read(data):
    s = int(input('Enter starting row : '))
    e = int(input('No of rows : '))
    print(data[s-1:(s-1)+e])
    return data
    

def update(data):
    while(True):
        i = int(input('Enter row number to update: '))
        print(data.iloc[[i-1]])
        col = input('\nEnter column name to update (Type the exact spelling as in the db): ')
        val = input('Enter Value for '+col+': ')
        data.at[i-1,col]=val
        print('Column Updated')
        c = input('Update more?')
        if c[:1].upper()=='Y': 
            continue
        else:
            return data
            break
    
def delete(data):
    r = int(input('Enter row number to delete: '))
    data = data.drop(r-1)
    print('\n',data.iloc[r-1])
    print('Row deleted')
    return data


if __name__ == "__main__":
    main()
```
